```python
import pandas as pd
exchange_rate_data = pd.read_csv("EXUSUK.csv")
```

```python
exchange_rate_data['DATE'] = pd.to_datetime(exchange_rate_data['DATE'])
exchange_rate_data.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(exchange_rate_data['DATE'],exchange_rate_data['EXUSUK'])
plt.xticks(rotation = 45)
plt.xlabel('Month')
plt.ylabel('Exchange Rate')
plt.title('Monthly U.S./U.K. Exchange Rate, 1975-2018')
plt.show()
```

```python
import math

exchange_rate_data['ln_EXUSUK'] = exchange_rate_data['EXUSUK'].map(lambda x:math.log(x))
exchange_rate_data.head()
```

```python
exchange_rate_data['diff']  = exchange_rate_data['ln_EXUSUK'] - exchange_rate_data['ln_EXUSUK'].shift(1)
exchange_rate_data.head()
```

```python
m = exchange_rate_data['diff'].mean()
print("Mean：%.2f%%"  %(m*100))
s = exchange_rate_data['diff'].std()
print("Standard Deviation：%.2f%%"  %(s*100))
```
